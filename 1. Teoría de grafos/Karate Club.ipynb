## Karate Club 2

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.karate_club_graph()
    
plt.figure(3,figsize=(6,6)) 
nx.draw_circular(G, with_labels=True, node_color ='#87E5DF')
plt.savefig('Figuras/KarateClub.png')
plt.show()

nEdges=G.number_of_edges()
nNodes=G.number_of_nodes()
```

```python
print(nx.is_connected(G))
```

### Matriz

```python
M=nx.to_numpy_matrix(G)
h=0
for i in M:
    print(h,i)
    h+=1
```

### Degree 

```python
degree=dict(G.degree())
degree
```

```python
avgDegree1=int(sum(degree.values())/2)
avgDegree1
```

```python
avgDegree2=nEdges
avgDegree2
```

### Connection Density

```python
k=2*nEdges/(nNodes*(nNodes-1))
k
```

### Average Shortest Path

```python
avgSPath1=nx.average_shortest_path_length(G)
avgSPath1
```

```python
#Creo que debe ser conexo, de lo contrario, "nNodes-1" estara mal
s=0
for i in G.nodes():
    m=sum(nx.shortest_path_length(G,i).values())/(nNodes-1)
    s=s+m
    
avgSPath2=s/nNodes
avgSPath2 
```

### Local Clustering Coefficient

```python
LCC=nx.clustering(G,12)
LCC
```

### Average Local Clustering Coefficient

```python
avgLCC=nx.average_clustering(G)
avgLCC
```

### Transitivity

```python
trans=nx.transitivity(G)
trans
```

```python
print("Degree: ", avgDegree2)
print("Connection Density: ", k)
print("Average Shortest Path: ", avgSPath1)
print("Average Local Clustering Coefficient: ", avgLCC)
print("Transitivity: ", trans)
```
